## How it's used in the vesicle viewer project

#### parameters
The parameters used by lmfit are named fit_parameters, and are packed as a Parameters object. 
In the parameters object are the following:
- shared parameters
    - chain volume (Vc)
    - headgroup volume (Vh)
    - terminal methyl volume (Vt)
- unknown parameters
    - area per lipid (Al)
    - headgroup thickness (Dh)
- smearing factor (sigma)
- separated form factors
    - average vesicle radius
    - relative size (polydispersity)
    
^ if there are multiple datasets, all these values will be the same. 
These following parameters are unique for each dataset:
- b values
    - chain b value (bc)
    - headgroup b value (bh)
    - terminal methyl b value (bt)
    - water b value (bw)
        - can either be electron density or scattering density?
- this is not a b value?
    - combined water volume (Vw)
- Tweaks to the graph, per dataset
    - scale
    - background (no, not the color or image of the background of the graph)

With these parameters, the fit is then minimized using lmfit and a custom objective function.

#### Objective function
The objective function unpacks the params in fit_parameters and calculates the residual using the defined mathematical model. 

The math (sym_model) only uses the fit_parameters, not the data (it only uses data.id to get the per-data-set values such as b values). It only uses the q value from data. 

Some values that are used outside of the sym model are the intensity value, and unpacked error correction values. 

The function returns a combined residual array for all datasets calculated.

```python
import matplotlib.pyplot as plt
import numpy as np

from lmfit.models import LinearModel
```

```python
# Step 1: random set of values to use lmfit (in vesicle-viewer these values are in a database)

values = np.random.randint(100, size=(10))
indices = np.arange(len(values))

print(values)
print(indices)
```

```python
# Step 2: place these values in lmfit, and fit using a given model

model = LinearModel()
params = model.guess(values, x=indices)
res = model.fit(values, params, x=indices)
```

```python
# Step 3: plot data (optional (: )
res.plot_fit()
plt.show()
```
